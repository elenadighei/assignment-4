```python
def degrees(n, edges):
    D = [0] * n
    for u,v in edges:
        D[u-1] += 1
        D[v-1] +=1
    return D

with open('/home/elenadg04/PoC2/rosalind_deg (1).txt', 'r') as file:
        n, m = map(int, file.readline().split())
        edges = [tuple(map(int, line.split())) for line in file.readlines()]

result = degrees(n, edges)
print(" ".join(map(str, result)))    
```

2. **Ddeg**

```python
def sum_neighbors(n, edges):
    D = [0] * n
    for u,v in edges:
        D[u-1] += 1
        D[v-1] +=1

    sum_of_neighbors_d = [0] * n 
    for u, v in edges:
        sum_of_neighbors_d[u-1] += D[v-1]
        sum_of_neighbors_d[v-1] += D[u-1]
    
    return sum_of_neighbors_d

with open('/home/elenadg04/PoC2/rosalind_ddeg.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        edges = [tuple(map(int, line.split())) for line in file.readlines()]

result = sum_neighbors(n, edges)
print(" ".join(map(str, result)))    
```

3. **Cc**

```python
def dfs(v, graph, visited):
    visited[v] = True
    for neighbor in graph[v]:
        if not visited[neighbor]:
            dfs(neighbor, graph, visited)

def connected_components(n, edges):
    graph = [[] for _ in range(n)]
    for u, v in edges:
        graph[u-1].append(v-1)  
        graph[v-1].append(u-1)  
    
    visited = [False] * n
    num_components = 0
    
    for i in range(n):
        if not visited[i]:
            dfs(i, graph, visited)
            num_components += 1
    
    return num_components

with open('/home/elenadg04/PoC2/rosalind_cc.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        edges = [tuple(map(int, line.split())) for line in file.readlines()]

result = connected_components(n, edges)
print(result)
```

4. **Bfs**

```python
def bfs(n, edges):
    graph = [[] for _ in range(n)]
    for u, v in edges:
        graph[u-1].append(v-1)

    distance = [-1] * n
    distance[0] = 0
    queue = [0]

    while queue:
        u = queue.pop(0)

        for v in graph[u]:
            if distance[v] == -1:  
                distance[v] = distance[u] + 1  
                queue.append(v)  
    
    return distance

with open('/home/elenadg04/PoC2/rosalind_bfs.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        edges = [tuple(map(int, line.split())) for line in file.readlines()]

result = bfs(n, edges)
print(" ".join(map(str, result)))
```

5. **Bip**

```python
def is_bipartite(n, edges):
    graph = {i: [] for i in range(1, n + 1)}
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    
    color = [-1] * (n + 1) 

    for start in range(1, n + 1):
        if color[start] == -1:  
            color[start] = 0
            queue = [start]
            while queue:
                node = queue.pop(0)
                for neighbor in graph[node]:
                    if color[neighbor] == -1:  
                        color[neighbor] = 1 - color[node]  
                        queue.append(neighbor)
                    elif color[neighbor] == color[node]:  
                        return -1
    return 1  

def process_input_bipartite(filename):
    with open(filename, 'r') as f:
        lines = [line.strip() for line in f if line.strip()]  
        k = int(lines[0])  
        results = []
        index = 1
        for _ in range(k):
            n, m = map(int, lines[index].split())  
            index += 1
            edges = []
            for _ in range(m):
                u, v = map(int, lines[index].split())
                edges.append((u, v))
                index += 1
            
            result = is_bipartite(n, edges)
            results.append(result)
        return results


filename = "rosalind_bip (5).txt"
results = process_input_bipartite(filename)
print(" ".join(map(str, results)))
```

6. **Dij**

```python
import heapq

def dij(n, edges):
    graph = {i: [] for i in range(1, n + 1)}
    for u, v, w in edges:
        graph[u].append((v, w))

    dist = [float('inf')] * (n + 1)
    dist[1] = 0
    pq = [(0, 1)] 

    while pq:
        current_dist, u = heapq.heappop(pq)
        if current_dist > dist[u]:
            continue

        for v, weight in graph[u]:
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                heapq.heappush(pq, (dist[v], v))

    return [dist[i] if dist[i] != float('inf') else -1 for i in range(1, n + 1)]

with open('/home/elenadg04/PoC2/rosalind_dij.txt', 'r') as file:
        n, m = map(int, file.readline().split())
        edges = [tuple(map(int, line.split())) for line in file.readlines()]

result = dij(n, edges)
print(" ".join(map(str, result)))                
```

7. **Dag**

```python
def is_acyclic(n, edges):
    adj = [[] for _ in range(n + 1)]
    in_degree = [0] * (n + 1)
    
    for u, v in edges:
        adj[u].append(v)
        in_degree[v] += 1
    
    queue = []
    
    for i in range(1, n + 1):
        if in_degree[i] == 0:
            queue.append(i)
    
    processed_count = 0
    
    while queue:
        node = queue.pop(0)
        processed_count += 1
        
        for neighbor in adj[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)
    
    if processed_count == n:
        return 1
    else:
        return -1

def main():
    with open('/home/elenadg04/PoC2/rosalind_dag (7).txt', 'r') as file:
        k = int(file.readline().strip())
        result = []
        
        for graph_index in range(k):
            line = file.readline().strip()
            while not line:
                line = file.readline().strip()
            
            n, m = map(int, line.split())
            
            edges = []
            for edge_index in range(m):
                edge_line = file.readline().strip()
                if edge_line:
                    edges.append(tuple(map(int, edge_line.split())))
            
            result.append(is_acyclic(n, edges))
        
        print(' '.join(map(str, result)))

main()

```

8. **Bf**

```python
def bellman(n, edges):
    D = [float('inf')] * n
    D[0] = 0

    for _ in range(n - 1):
        for u, v, w in edges:
            if D[u] != float('inf') and D[u] + w < D[v]:
                D[v] = D[u] + w

    result = []
    for dist in D:
        if dist == float('inf'):
            result.append('x')
        else:
            result.append(dist)
    
    return result

with open('/home/elenadg04/PoC2/rosalind_bf.txt', 'r') as file:
    n, m = map(int, file.readline().split())
    edges = []
    for line in file:
        u, v, w = map(int, line.strip().split())
        edges.append((u - 1, v - 1, w))

result = bellman(n, edges)
print(" ".join(map(str, result)))
```

9. **Cte**

```python
import heapq

def dijkstra(graph, source, n):
    dist = [float('inf')] * (n + 1)
    dist[source] = 0
    pq = [(0, source)]
    
    while pq:
        current_dist, u = heapq.heappop(pq)
        
        if current_dist > dist[u]:
            continue
        
        for v, weight in graph[u]:
            new_dist = current_dist + weight
            if new_dist < dist[v]:
                dist[v] = new_dist
                heapq.heappush(pq, (new_dist, v))
    
    return dist

def find_shortest_cycle(graph, n, u1, v1, w1):
    dist_from_v1 = dijkstra(graph, v1, n)
    
    if dist_from_v1[u1] == float('inf'):
        return -1
    
    return dist_from_v1[u1] + w1

def solve():
    with open("rosalind_cte (4).txt", "r") as file:
        k = int(file.readline().strip())
        
        results = []
        
        for _ in range(k):
            n, m = map(int, file.readline().strip().split())
            graph = [[] for _ in range(n + 1)]
            edges = []
            
            for i in range(m):
                u, v, w = map(int, file.readline().strip().split())
                graph[u].append((v, w))
                edges.append((u, v, w))
            
            u1, v1, w1 = edges[0]
            
            result = find_shortest_cycle(graph, n, u1, v1, w1)
            results.append(str(result))
        
        print(" ".join(results))

solve()
```

10. **Nwc**

```python
def bellman_ford(n, edges):
    dummy_edges = edges + [(0, i, 0) for i in range(1, n + 1)]
    
    
    dist = [float('inf')] * (n + 1)
    dist[0] = 0  

    for _ in range(n):
        updated = False
        for u, v, w in dummy_edges:
            if dist[u] != float('inf') and dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                updated = True

        if not updated:
            break

    for u, v, w in dummy_edges:
        if dist[u] != float('inf') and dist[u] + w < dist[v]:
            return 1  

    return -1  


def process_input(filename):
    with open(filename, 'r') as f:
        k = int(f.readline().strip())  
        
        results = []
        for _ in range(k):
            n, m = map(int, f.readline().strip().split()) 
            edges = []
            for _ in range(m):
                u, v, w = map(int, f.readline().strip().split())
                edges.append((u, v, w))
            
            
            result = bellman_ford(n, edges)
            results.append(result)
        
        return results


filename = "/home/elenadg04/PoC2/rosalind_nwc (11).txt"
results = process_input(filename)
print(" ".join(map(str, results)))

```

11. **Sdag**

```python
def topological_sort(n, edges):
    in_degree = [0] * (n + 1)
    graph = {i: [] for i in range(1, n + 1)}

    for u, v, w in edges:
        graph[u].append((v, w))
        in_degree[v] += 1

    queue = [i for i in range(1, n + 1) if in_degree[i] == 0]
    topo_order = []

    while queue:
        node = queue.pop(0)  
        topo_order.append(node)
        for neighbor, _ in graph[node]:
            in_degree[neighbor] -= 1
            if in_degree[neighbor] == 0:
                queue.append(neighbor)

    return topo_order

def shortest_paths_dag(n, edges):
    graph = {i: [] for i in range(1, n + 1)}
    for u, v, w in edges:
        graph[u].append((v, w))

    topo_order = topological_sort(n, edges)
    dist = [float('inf')] * (n + 1)
    dist[1] = 0

    for u in topo_order:
        if dist[u] < float('inf'):
            for v, w in graph[u]:
                if dist[u] + w < dist[v]:
                    dist[v] = dist[u] + w

    return [0 if i == 1 else (dist[i] if dist[i] < float('inf') else 'x') for i in range(1, n + 1)]

def process_input_dag(filename):
    with open(filename, 'r') as f:
        n, m = map(int, f.readline().strip().split())
        edges = []
        for _ in range(m):
            u, v, w = map(int, f.readline().strip().split())
            edges.append((u, v, w))
        return shortest_paths_dag(n, edges)


filename = "rosalind_sdag.txt"
result = process_input_dag(filename)
print(" ".join(map(str, result)))
```

14. **Long**

```python
from Bio import SeqIO

DNA = []
for record in SeqIO.parse("/home/elenadg04/PoC2/rosalind_long.txt", "fasta"):
    DNA.append(str(record.seq))

def overlap(a, b):
    max_overlap = 0
    for i in range(1, len(a)):
        if b.startswith(a[i:]):  
            max_overlap = len(a) - i
            break
    return max_overlap   

def merge(a, b, overlap_len):
    return a + b[overlap_len:]


def genome_reconstruction(DNA):
    while len(DNA) > 1:
        max_overlap = -1
        best_pair = None
        best_a, best_b = None, None

        
        for i in range(len(DNA)):
            for j in range(len(DNA)):
                if i != j:
                    olap = overlap(DNA[i], DNA[j])
                    if olap > max_overlap:
                        max_overlap = olap
                        best_pair = (i, j)
                        best_a, best_b = DNA[i], DNA[j]

        
        if best_pair:
            i, j = best_pair
            merged = merge(best_a, best_b, max_overlap)
            DNA = [DNA[k] for k in range(len(DNA)) if k != i and k != j]  
            DNA.append(merged)

    return DNA[0]        

result = genome_reconstruction(DNA)
print(result)    
```
